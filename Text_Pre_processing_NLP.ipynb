```python
# Import required libraries
import nltk
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

```python
# Download required NLTK data
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
text = """
Artificial Intelligence (AI) is transforming the world! It is used in healthcare, finance, education, and transportation.
Machines can learn from data, analyze patterns, and make decisions. However, AI also raises questions: Will machines replace humans?

Natural Language Processing (NLP) is a branch of AI. It helps computers understand human language; for example, chatbots, translation systems, and voice assistants.
NLP techniques include tokenization, stemming, lemmatization, and stop-word removal. These techniques improve text analysis!
"""
```

```python
print("\nOriginal Text:\n")
print(text)
```

```python
# Remove Punctuation

text_no_punct = text.translate(str.maketrans('', '', string.punctuation))

print("\nText After Removing Punctuation:\n")
print(text_no_punct)
```

```python
# Tokenization

tokens = word_tokenize(text_no_punct)

print("\nTokens:\n")
print(tokens)
```

```python
#Stop Word Removal

stop_words = set(stopwords.words('english'))

filtered_words = [word for word in tokens if word.lower() not in stop_words]

print("\nAfter Stop Word Removal:\n")
print(filtered_words)
```

```python
#Stemming

stemmer = PorterStemmer()

stemmed_words = [stemmer.stem(word) for word in filtered_words]

print("\nAfter Stemming:\n")
print(stemmed_words)
```

```python
# Lemmatization

lemmatizer = WordNetLemmatizer()

lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]

print("\nAfter Lemmatization:\n")
print(lemmatized_words)
```
